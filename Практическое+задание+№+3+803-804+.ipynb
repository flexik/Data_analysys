**Шаг 1.** Импортируйте библиотеки. Подключите источник данных http://books.toscrape.com/catalogue/page-1.html

```python
import requests              #импортирую необходимые библиотеки
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup

print (np.__version__)
print (pd.__version__)
print (requests.__version__)
```

**Шаг 2.** С помощью библиотеки BeautifulSoup найдите HTML теги, в которых находится информация по товарам (Название книги, цена).

```python
url = 'http://books.toscrape.com/catalogue/page-1.html'
product = list() #сюда будем записывать товары и цены

tek_page = 1
pages = 5 # задаем число больше номера первой страницы, для старта цикла

while tek_page <= pages:                                # перебираем циклом 'while' 5 страниц, извлекая оттуда циклом 'for'
    newUrl = url.replace('1', str(tek_page))            # теги с названием и ценой и помещаем их в список 'product'. 
    response = requests.get(newUrl)
    soup = BeautifulSoup(response.text, 'html.parser')
    items = soup.find_all('li', {'class':'col-xs-6 col-sm-4 col-md-3 col-lg-3'})
    for n, i in enumerate(items):
        name = i.find('h3').get_text().strip()
        price = i.find('p', {'class':'price_color'}).get_text().strip()[1:]
        print(f'{n}. {name} {price}')
        product.append(
            {'name' : name,
             'price' : price
            })
    tek_page +=1
```

**Шаг 3.** Выведите результат в DataFarme

```python
df = pd.DataFrame(data=product).rename(columns = {'name' : 'Название книги', 'price' : 'Цена'}) # передаю данные в df + переименовываю колонки, в соотв. с заданием.
df.head(30)        # вывожу первые 30 строк для наглядности            
```

**Шаг 4.** Сохраните DataFarme в файл csv.

```python
FILE_NAME = "practice_3.csv"     
df.to_csv(FILE_NAME)
```
